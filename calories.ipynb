# 1. Linguistic variables and Membership Functions (MFs)

### Helper functions for MFs

```python
def trapmf(x, abcd):
    a, b, c, d = abcd
    if a > b: raise ValueError('a > b')
    if b > c: raise ValueError('b > c') # Allow b == c for triangular
    if c > d: raise ValueError('c > d')
        
    y = np.zeros_like(x if isinstance(x, np.ndarray) else [x])
    
    # Handle scalar input
    if not isinstance(x, np.ndarray):
        x = np.array([x])
        y = np.zeros_like(x, dtype=float)
        single_value = True
    else:
        y = np.zeros_like(x, dtype=float)
        single_value = False

    # Left slope
    if a != b:
        idx = np.logical_and(x >= a, x < b)
        y[idx] = (x[idx] - a) / (b - a)

    # Top part
    idx = np.logical_and(x >= b, x <= c)
    y[idx] = 1

    # Right slope
    if c != d:
        idx = np.logical_and(x > c, x <= d)
        y[idx] = (d - x[idx]) / (d - c)

    return y[0] if single_value else y
        
    return y[0] if isinstance(y, np.ndarray) and y.size == 1 else y

def trimf(x, abc):
    a, b, c = abc
    if a > b: raise ValueError('a > b in trimf')
    if b > c: raise ValueError('b > c in trimf')
    return trapmf(x, [a, b, b, c])
```

### Input: BMI (kg/m)

```python
def bmi_underweight(bmi):
    return trapmf(bmi, [15, 15, 17, 18.5])

def bmi_normal(bmi):
    return trimf(bmi, [17, 21.75, 25])

def bmi_overweight(bmi):
    return trimf(bmi, [25, 27.5, 30])

def bmi_obese(bmi):
    return trapmf(bmi, [30, 35, 50, 50])

# Visualize BMI MFs
x_bmi_vis = np.arange(10, 55, 0.1)
plt.figure(figsize=(10, 6))
plt.plot(x_bmi_vis, bmi_underweight(x_bmi_vis), label='Underweight')
plt.plot(x_bmi_vis, bmi_normal(x_bmi_vis), label='Normal')
plt.plot(x_bmi_vis, bmi_overweight(x_bmi_vis), label='Overweight')
plt.plot(x_bmi_vis, bmi_obese(x_bmi_vis), label='Obese')
plt.title('Membership Functions for BMI')
plt.xlabel('BMI (kg/m)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()
```

# 2. Fuzzification

```python
bmi_mf_terms = {
    'Underweight': bmi_underweight,
    'Normal': bmi_normal,
    'Overweight': bmi_overweight,
    'Obese': bmi_obese
}

bodyfat_m_mf_terms = { # Men
    'Lean_M': bfm_lean, 
    'Normal_M': bfm_normal, 
    'High_M': bfm_high
}

bodyfat_f_mf_terms = { # Women
    'Lean_F': bfw_lean, 
    'Normal_F': bfw_normal, 
    'High_F': bfw_high
}

age_mf_terms = {
    'Young': age_young,
    'Middle': age_middle,
    'Senior': age_senior
}

calorie_output_mamdani_terms = {
    'Low': calorie_low_mf,
    'Medium': calorie_medium_mf,
    'High': calorie_high_mf
}
```

```python
def fuzzify_inputs(bmi, bodyfat, sex, age):
    fuzz_bmi = {term: mf(bmi) for term, mf in bmi_mf_terms.items()}
    
    if sex == 'M':
        fuzz_bf = {term: mf(bodyfat) for term, mf in bodyfat_m_mf_terms.items()}
    else: # sex == 'F'
        fuzz_bf = {term: mf(bodyfat) for term, mf in bodyfat_f_mf_terms.items()}
        
    fuzz_age = {term: mf(age) for term, mf in age_mf_terms.items()}
    
    return fuzz_bmi, fuzz_bf, fuzz_age

# Example fuzzification
example_bmi_cal = 23
example_bf_cal = 20
example_sex_cal = 'F'
example_age_cal = 35

f_bmi_cal, f_bf_cal, f_age_cal = fuzzify_inputs(example_bmi_cal, example_bf_cal, example_sex_cal, example_age_cal)
print(f"Fuzzified BMI for {example_bmi_cal}: {f_bmi_cal}")
print(f"Fuzzified BodyFat ({example_sex_cal}) for {example_bf_cal}%: {f_bf_cal}")
print(f"Fuzzified Age for {example_age_cal} years: {f_age_cal}")
```

```python
def apply_calorie_rules(fuzz_bmi, fuzz_bf, fuzz_age, sex_input):
    aggregated_mamdani_strengths = {term: 0.0 for term in calorie_output_mamdani_terms}
    sugeno_rule_outputs = []
    active_rules_details_list = []

    for rule_sex, bmi_term, bf_term, age_term, calorie_consequent_term in calorie_rules:
        if rule_sex != sex_input:
            continue

        bmi_strength = fuzz_bmi.get(bmi_term, 0)
        bf_strength = fuzz_bf.get(bf_term, 0)
        age_strength = fuzz_age.get(age_term, 0)
        
        rule_strength = min(bmi_strength, bf_strength, age_strength)
        
        if rule_strength > 0:
            active_rules_details_list.append({
                'Sex': rule_sex,
                'BMI Term': bmi_term, 'BMI Strength': bmi_strength,
                'BF Term': bf_term, 'BF Strength': bf_strength,
                'Age Term': age_term, 'Age Strength': age_strength,
                'Calorie Consequent': calorie_consequent_term,
                'Rule Strength': rule_strength
            })
            
            aggregated_mamdani_strengths[calorie_consequent_term] = max(
                aggregated_mamdani_strengths[calorie_consequent_term], rule_strength
            )
            
            sugeno_consequent_value = calorie_sugeno_constants[calorie_consequent_term]
            sugeno_rule_outputs.append((rule_strength, sugeno_consequent_value))
            
    return aggregated_mamdani_strengths, sugeno_rule_outputs, pd.DataFrame(active_rules_details_list)
```

```python
# Example rule application for Calories
agg_mamdani_cal, sug_outputs_cal, active_rules_cal_df = apply_calorie_rules(f_bmi_cal, f_bf_cal, f_age_cal, example_sex_cal)
print("Aggregated Mamdani Strengths (Calories):", agg_mamdani_cal)
print("\nSugeno Rule Outputs (Calories - strength, consequent_value):", sug_outputs_cal)
print("\nActive Rules Details (Calories):")
if not active_rules_cal_df.empty:
    display(active_rules_cal_df)
else:
    print("No rules activated for the given inputs.")
```

# 4. Defuzzification

### 4.1 Mamdani (Centroid)

```python
calorie_universe_defuzz = np.arange(1200, 3501, 10) # Universe for calorie output for defuzzification

def defuzzify_mamdani_centroid_calorie(aggregated_strengths, output_universe):
    numerator = 0
    denominator = 0
    aggregated_mf_plot_values = np.zeros_like(output_universe, dtype=float)

    for i, cal_val in enumerate(output_universe):
        max_membership_for_cal_val = 0
        for calorie_term, strength in aggregated_strengths.items():
            if strength > 0:
                clipped_value = min(calorie_output_mamdani_terms[calorie_term](cal_val), strength)
                max_membership_for_cal_val = max(max_membership_for_cal_val, clipped_value)
        aggregated_mf_plot_values[i] = max_membership_for_cal_val
        numerator += cal_val * max_membership_for_cal_val
        denominator += max_membership_for_cal_val
        
    # Optional: Plotting the aggregated MF
    # plt.figure()
    # plt.plot(output_universe, aggregated_mf_plot_values, label='Aggregated Output MF (Calories)')
    # plt.title('Aggregated Output MF - Mamdani (Calories)')
    # plt.xlabel('Calories (kcal)')
    # plt.ylabel('Membership Degree')
    # plt.fill_between(output_universe, aggregated_mf_plot_values, alpha=0.4)
    # plt.legend()
    # plt.grid(True)
    # plt.show()

    if denominator == 0:
        return 0 
    return numerator / denominator
```

```python
# Example Mamdani defuzzification for Calories
mamdani_calorie_prediction = defuzzify_mamdani_centroid_calorie(agg_mamdani_cal, calorie_universe_defuzz)
print(f"Predicted Calories (Mamdani Centroid): {mamdani_calorie_prediction:.2f} kcal")
```

### 4.2 Sugeno (Weighted Average)

```python
def defuzzify_sugeno_weighted_average_calorie(sugeno_rule_outputs_list):
    # sugeno_rule_outputs_list is a list of (strength, consequent_value)
    numerator = 0
    denominator = 0
    
    if not sugeno_rule_outputs_list:
        return 0
        
    for strength, consequent_value in sugeno_rule_outputs_list:
        numerator += strength * consequent_value
        denominator += strength
        
    if denominator == 0:
        return 0
        
    return numerator / denominator
```

```python
# Example Sugeno defuzzification for Calories
sugeno_calorie_prediction = defuzzify_sugeno_weighted_average_calorie(sug_outputs_cal)
print(f"Predicted Calories (Sugeno Weighted Average): {sugeno_calorie_prediction:.2f} kcal")
```

# 5. Evaluation

```python
mamdani_calorie_predictions_eval = []
sugeno_calorie_predictions_eval = []
actual_calorie_values_eval = []

for index, row in df_cleaned.iterrows():
    bmi_val = row['BMI']
    bodyfat_val = row['BodyFat']
    sex_val = row['Sex']
    age_val = row['Age']
    actual_calorie = row['Calories_kcal']
    
    f_bmi, f_bf, f_age = fuzzify_inputs(bmi_val, bodyfat_val, sex_val, age_val)
    agg_mamdani_str, sug_outputs_list, _ = apply_calorie_rules(f_bmi, f_bf, f_age, sex_val)
    
    mamdani_pred = defuzzify_mamdani_centroid_calorie(agg_mamdani_str, calorie_universe_defuzz)
    sugeno_pred = defuzzify_sugeno_weighted_average_calorie(sug_outputs_list)
    
    mamdani_calorie_predictions_eval.append(mamdani_pred)
    sugeno_calorie_predictions_eval.append(sugeno_pred)
    actual_calorie_values_eval.append(actual_calorie)

df_cleaned['Calorie_Mamdani_Predicted'] = mamdani_calorie_predictions_eval
df_cleaned['Calorie_Sugeno_Predicted'] = sugeno_calorie_predictions_eval

print("\n--- Sample of Actual vs Predicted Calories (with predictions) ---")
print(df_cleaned[['Calories_kcal', 'Calorie_Mamdani_Predicted', 'Calorie_Sugeno_Predicted']].head())
```

```python
# F1 Score Evaluation for Calories
calorie_bins_edges_eval = [0, 1900, 2550, float('inf')] # Based on Low, Medium, High ranges
calorie_labels_eval = [0, 1, 2] # Representing Low, Medium, High

actual_cal_binned_eval = pd.cut(df_cleaned['Calories_kcal'], bins=calorie_bins_edges_eval, labels=calorie_labels_eval, include_lowest=True, duplicates='drop')
mamdani_cal_binned_eval = pd.cut(df_cleaned['Calorie_Mamdani_Predicted'], bins=calorie_bins_edges_eval, labels=calorie_labels_eval, include_lowest=True, duplicates='drop')
sugeno_cal_binned_eval = pd.cut(df_cleaned['Calorie_Sugeno_Predicted'], bins=calorie_bins_edges_eval, labels=calorie_labels_eval, include_lowest=True, duplicates='drop')

# Drop NA values that might result from binning if predictions are outside the universe or due to sparse data in bins
valid_indices_mamdani_cal = actual_cal_binned_eval.notna() & mamdani_cal_binned_eval.notna()
valid_indices_sugeno_cal = actual_cal_binned_eval.notna() & sugeno_cal_binned_eval.notna()

if valid_indices_mamdani_cal.sum() > 0 and len(np.unique(actual_cal_binned_eval[valid_indices_mamdani_cal])) > 1:
    f1_mamdani_cal = metrics.f1_score(actual_cal_binned_eval[valid_indices_mamdani_cal], mamdani_cal_binned_eval[valid_indices_mamdani_cal], average='macro', zero_division=0)
    print(f"\nF1 Score (Mamdani Calories - Binned): {f1_mamdani_cal:.4f}")
else:
    print("\nCould not calculate Mamdani F1 score for Calories due to insufficient data after binning/NA removal or only one class present.")

if valid_indices_sugeno_cal.sum() > 0 and len(np.unique(actual_cal_binned_eval[valid_indices_sugeno_cal])) > 1:
    f1_sugeno_cal = metrics.f1_score(actual_cal_binned_eval[valid_indices_sugeno_cal], sugeno_cal_binned_eval[valid_indices_sugeno_cal], average='macro', zero_division=0)
    print(f"F1 Score (Sugeno Calories - Binned): {f1_sugeno_cal:.4f}")
else:
    print("\nCould not calculate Sugeno F1 score for Calories due to insufficient data after binning/NA removal or only one class present.")
```
